<h2>Background and Motivation</h2>
In 2017, new solar PV capacity grew by 50% globally, reaching over 74 GW and surpassing net growth in coal for the first time.  The U.S. remained the second largest growth market for renewables, after China.  It is known that the efficiency of solar panels is heavily dependent on operating temperature.  Commonly installed mono-and polycrystalline solar cells experience an efficiency loss of 0.5% for each degree above 25oC and many solar modules can reach 60 oC in the field.

All three team members study engineering and two currently work with renewable energy systems (wind and solar).  Brooke currently works in the Wind Energy & Turbulence Lab in Mechanical Engineering and part of the lab’s research focuses on developing a better understanding of the physical interactions between solar farms and the atmospheric boundary layer, both from a numerical and experimental perspective.  Specifically, the research group aims to develop a model for the convective heat transfer coefficient ‘h’ of solar modules within solar farms.

Once the heat transfer between solar modules and the atmosphere is understood and ‘h’ can be predicted, engineering solutions can be developed and evaluated to manipulate ‘h’ in a way that reduces the operating temperature of solar modules, increasing efficiency and thus, power output.


<h2>Project Objectives</h2>
The primary objective of this project is to develop a model for the convective heat transfer coefficient ‘h’ of solar modules within solar farms from experimental data.  In addition, we aim to address the following questions about the dataset:
1.	Is there a successful model for ‘h’ that depends only on data is typically measured near solar power plants (wind speed, wind direction, ambient temperature, radiation, and relative humidity)? 
a.	If not, which other variables are necessary to predict ‘h’ accurately?
2.	Of the experimental variables, what are the strongest contributors to ‘h’?
3.	What other trends and relationships do we see among the atmospheric variables such as wind speed, wind direction, stability, and time?

<h2>Data Processing</h2>
Much of the data for these variables is currently stored in a .csv file for the entire experiment duration, however, the sonic anemometer data is scattered among numerous .dat files, one for each day consisting of ~1.7 million lines each.  Extensive cleaning, reshaping, and homogenizing of the data will be required.  From the experimental data, 3 “turbulence” variables can be calculated, which may be important contributors to the model for ‘h’.

The Python Pandas module will be used to import and clean the data, specifically with the read_csv function. Modules such as numpy and sklearn may be used in reshaping and homogenizing the data as well.


<h2>Exploratory Analysis</h2>
For exploratory analysis, the processed and cleaned data will be used to determine the correlations
between the predictor variables and the target variables. This will give us a good idea of which variables
appear to be significantly impacting the value of h and which can be ignored. We will also use these
predictor variables in a well-studied physical model of a plate in wind, an image of which is shown

There are several equations and empirical correlations that have been developed for this simplified
model. We will use these with our experimental data to determine the accuracy of this model, and if it
can reliably predict the heat transfer coefficient.

<h2>Analysis Methodology</h2>
The full analysis will depend on whether the scarecrow model for a flat plate in parallel flow is sufficient
for our data. If it is not (the likely scenario), then we will use the data we have to work backwards and
come up with the best empirical correlation that accounts for the variance of predictor variables that
the simplified model does not include. We will test this new model and see the improvement in accuracy
in determining h based on the conditions, which can then potentially be used to predict and minimize
unnecessary heat losses in the field.

If the simplified model does obtain a reasonably good degree of predictive accuracy, then our data
analysis will instead focus on classification. Because the treatment of the plate under laminar and
turbulent flow is different, we would aim to create a SVM that distinguishes the two types of flow based
on predictor variables. We will also classify the stability of the atmosphere, which is typically assumed as
unstable during the day and stable during the night, but a classifier based on these conditions could be
very beneficial for other analyses.
