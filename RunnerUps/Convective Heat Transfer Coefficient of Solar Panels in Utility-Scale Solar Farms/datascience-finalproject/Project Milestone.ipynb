<h2>Data Acquisition</h2>

The data was collected during the summer of 2017 at a solar PV plant at the Denver Federal Center.  The data collection sensors and methods are shown in the diagram below.
![firstdatapic.png](firstdatapic.png)
Most of the data was obtained as .csv files, which includes wind speed and direction data as measured by mechanical "cup and vane" anemometers. However, the sonic wind anemometer, which obtained more detailed, accurate measurements of the wind speed required much more to use detailed in the preprocessing section. 

Why go through all this trouble to use the sonic anemometer data?  Unlike the traditional cup anemometer and vane devices to measure wind speed and direction, a sonic anemometer captures turbulence through sending ultrasonic pulses among 3 transducers.  Sonic anemometers capture all 3 dimensions of wind speed as well as rapid changes and gusts.  It is possible that these gusts and rapid changes in wind flow may impact the local 'h' for solar panels within solar farms.  For more information on sonic vs cup & vane anemometers, please visit https://www.windpowerengineering.com/simulation/ultrasonic-wind-sensors-or-cup-anemometers/.

It will be interesting to determine the relevance of the mechanical, more affordably-acquired wind speed data versus the sonic wind speed data.  It is possible that we determine that the more expensive sonic data is not worth the time and cost.

<h2>Data Preprocessing</h2>

The data in the .csv file was loaded using Pandas dataframes and scaled to prevent any one measure of something from being weighted improperly. However, the sonic anemometer data required lengthy preprocessing in order to be usable.  As mentioned in the project proposal, the sonic anemometer data was taken at 20Hz frequencey and is scattered among numerous .dat files, one for each day consisting of ~1.7 million lines each.  Additionally, the data consists of wind speed values in the x, y, z direction for three different heights on the weather tower.  This data needs to be reduced significantly and oriented to align with the "mean wind velocity" and normalized in order to be usable.

A former member of U of Utah's Wind Energy & Turbulence Lab developed "UTESpac" found at the link below that computes means, fluxes, variances and derived temperatures (potential temperature, virtual potential temperature) and stores the output in a MatLab structure or NetCDF file.
https://github.com/derek591/UTESpac

In order to use this code, I first needed to re-convert all the wind data from the raw binary format collected by Campbell Scientific dataloggers to a usable CSV format using the accompanying LoggerNet software.  This step is complete for a large subset of the experiment and will be completed for all missing data before the final stage of the project.

I then used UTESpac to reduce the data using 30-minute averages, re-orient the measurements, and store the data into a Matlab structure.  Using UTESpac required debugging of the code in Matlab due to slightly different filenames than what was expected by the code.  In this output format, the data from the entire experiment can be plotted so that we may select certain time chunks to use for analysis.

This Matlab structure, or a more manageable subset of the structure, can then be converted to Pandas dataframes in Python and treated similarly to the rest of the data.  This step is still to be completed, but I plan to try the two methods mentioned in this article (1. scipy loadmat fucntion or 2. mat4py loadmat function: https://stackoverflow.com/questions/38197449/matlab-data-file-to-pandas-dataframe.
